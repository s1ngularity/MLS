```python
from xml.dom import minidom
xmldoc = minidom.parse('train-data.xml')
instances = xmldoc.getElementsByTagName('instance')
```

```python
for exp in instances:
    print('',exp.getAttribute('id'),end='')
    print(").",end='')
    text = exp.getElementsByTagName('text')
    for t in text:
        print(t.childNodes[0].nodeValue)
        questions = exp.getElementsByTagName('question')
        for q in questions:
            print("\t",q.getAttribute('id'),"-",q.getAttribute('text'))
            answers = q.getElementsByTagName('answer')
            for a in answers:
                print('\t\t',a.getAttribute('id'),'-', a.getAttribute('text'))
```
